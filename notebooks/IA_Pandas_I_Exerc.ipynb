## Q3. Qual a média dos pagamentos (charges) efetuados?

```python
df['charges'].mean()
```

## Q4. Qual a média dos pagamentos (charges) dos 10 maiores pagamentos efetuados?

```python
df['charges']\
  .sort_values(ascending=False)\
  .reset_index(drop=True)[:10]\
  .mean()
```

## Q5. Calcule a distância interquartis dos pagamentos.

**Dica** IQR = Q3 - Q1



```python
df['charges'].describe()
```

```python
# Calcular o primeiro quartil (Q1)
q1 = df['charges'].quantile(0.25)

# Calcular o terceiro quartil (Q3)
q3 = df['charges'].quantile(0.75)

# Calcular a distância interquartil
interquartil_range = q3 - q1

print("Primeiro Quartil (Q1):", q1)
print("Terceiro Quartil (Q3):", q3)
print("Distância Interquartil:", interquartil_range)
```

## Q6. Empregue o valor IQR da questão anterior para verificar se existem *outliers* dos pagamentos.

**Dica** *outliers* são dados fora do intervalo $[Q1 - 1.5 IQR, Q3 + 1.5 IQR]$

```python
min_outlier = q1 - 1.5 * interquartil_range
max_outlier = q3 + 1.5 * interquartil_range

print(min_outlier, max_outlier)

# calculate outliers
outliers_inf =  df[(df['charges'] < min_outlier)]
outliers_sup =  df[(df['charges'] > max_outlier)]

print(len(outliers_inf), len(outliers_sup))
```
